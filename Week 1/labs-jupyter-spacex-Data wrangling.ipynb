Located at 22,236 miles (35,786 kilometers) above Earth's equator, this position is a valuable spot for monitoring weather, communications and surveillance. Because the satellite orbits at the same speed that the Earth is turning, the satellite seems to stay in place over a single longitude, though it may drift north to south,‚Äù NASA wrote on its Earth Observatory website <a  href="https://www.space.com/29222-geosynchronous-orbit.html" >[3] </a>.


* <b>SSO (or SO)</b>: It is a Sun-synchronous orbit  also called a heliosynchronous orbit is a nearly polar orbit around a planet, in which the satellite passes over any given point of the planet's surface at the same local mean solar time <a href="https://en.wikipedia.org/wiki/Sun-synchronous_orbit">[4] <a>.

We create a set of outcomes where the second stage did not land successfully:


```python
asds = set(landing_outcomes.keys()[[0,3]])
rtls = set(landing_outcomes.keys()[[2,7]])
ocean = set(landing_outcomes.keys()[[4,5]])
```

```python
# Define sets based on specific indices
asds_indices = [0, 3]
rtls_indices = [2, 7]
ocean_indices = [4, 5]

# Initialize counts for success and failure for each pair
asds_success_count = 0
asds_failure_count = 0

rtls_success_count = 0
rtls_failure_count = 0

ocean_success_count = 0
ocean_failure_count = 0

# Iterate through the landing outcomes and count successes and failures for each pair
for i, count in enumerate(landing_outcomes):
    if i in asds_indices:
        asds_success_count += count
    else:
        asds_failure_count += count

    if i in rtls_indices:
        rtls_success_count += count
    else:
        rtls_failure_count += count

    if i in ocean_indices:
        ocean_success_count += count
    else:
        ocean_failure_count += count

# Calculate total counts for each pair
asds_total_count = asds_success_count + asds_failure_count
rtls_total_count = rtls_success_count + rtls_failure_count
ocean_total_count = ocean_success_count + ocean_failure_count

# Calculate success and failure ratios for each pair
asds_success_ratio = asds_success_count / asds_total_count if asds_total_count != 0 else 0
asds_failure_ratio = asds_failure_count / asds_total_count if asds_total_count != 0 else 0

rtls_success_ratio = rtls_success_count / rtls_total_count if rtls_total_count != 0 else 0
rtls_failure_ratio = rtls_failure_count / rtls_total_count if rtls_total_count != 0 else 0

ocean_success_ratio = ocean_success_count / ocean_total_count if ocean_total_count != 0 else 0
ocean_failure_ratio = ocean_failure_count / ocean_total_count if ocean_total_count != 0 else 0

# Convert ratios to percentages with two decimal places
asds_success_percentage = asds_success_ratio * 100
asds_failure_percentage = asds_failure_ratio * 100

rtls_success_percentage = rtls_success_ratio * 100
rtls_failure_percentage = rtls_failure_ratio * 100

ocean_success_percentage = ocean_success_ratio * 100
ocean_failure_percentage = ocean_failure_ratio * 100

# Print formatted percentages
print("ASDS Success Percentage: {:.2f}%".format(asds_success_percentage))
print("ASDS Failure Percentage: {:.2f}%".format(asds_failure_percentage))

print("RTLS Success Percentage: {:.2f}%".format(rtls_success_percentage))
print("RTLS Failure Percentage: {:.2f}%".format(rtls_failure_percentage))

print("Ocean Success Percentage: {:.2f}%".format(ocean_success_percentage))
print("Ocean Failure Percentage: {:.2f}%".format(ocean_failure_percentage))
```

```python
# Set Seaborn style
sns.set_style("whitegrid")

# Define labels for the pairs
labels = ['ASDS', 'RTLS', 'Ocean']

# Define success and failure percentages
success_percentages = [asds_success_percentage, rtls_success_percentage, ocean_success_percentage]
failure_percentages = [asds_failure_percentage, rtls_failure_percentage, ocean_failure_percentage]

data = {
    'Pairs': labels * 2,
    'Percentage': success_percentages + failure_percentages,
    'Outcome': ['Success'] * len(labels) + ['Failure'] * len(labels)
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Pairs', y='Percentage', hue='Outcome', data=df, palette={'Success': 'green', 'Failure': 'red'})
plt.title('Success and Failure Percentage of Each Landing Type', fontsize=16)
plt.ylabel('Percentage', fontsize=14)
plt.xlabel('Landing Type', fontsize=14)
plt.yticks(fontsize=12)
plt.legend(title='Outcome', fontsize=12)

# Add percentage values on each bar
for p in ax.patches:
    height = p.get_height()
    if height != 0:  # Exclude bars with 0% success/failure
        ax.annotate('{:.2f}%'.format(height),
                    xy=(p.get_x() + p.get_width() / 2, height),
                    xytext=(0, 3), 
                    textcoords="offset points",
                    ha='center',
                    fontsize=10)

plt.tight_layout()
plt.show()
```

### TASK 4: Create a landing outcome label from Outcome column


Using the <code>Outcome</code>,  create a list where the element is zero if the corresponding  row  in  <code>Outcome</code> is in the set <code>bad_outcome</code>; otherwise, it's one. Then assign it to the variable <code>landing_class</code>:


```python
# landing_class = 0 if bad_outcome
# landing_class = 1 otherwise


landing_class = [0 if outcome in bad_outcomes else 1 for outcome in df['Outcome']]
# landing_class
```

This variable will represent the classification variable that represents the outcome of each launch. If the value is zero, the  first stage did not land successfully; one means  the first stage landed Successfully 


```python
df['Class']= landing_class
df[['Class']].head(8)
```
