## Exploratory Data Analysis 


First, let's read the SpaceX dataset into a Pandas dataframe and print its summary


```python
df=pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")

# If you were unable to complete the previous lab correctly you can uncomment and load this csv

# df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0701EN-SkillsNetwork/api/dataset_part_2.csv')

df.head(5)
```

First, let's try to see how the `FlightNumber` (indicating the continuous launch attempts.) and `Payload` variables would affect the launch outcome.

We can plot out the <code>FlightNumber</code> vs. <code>PayloadMass</code>and overlay the outcome of the launch. We see that as the flight number increases, the first stage is more likely to land successfully. The payload mass is also important; it seems the more massive the payload, the less likely the first stage will return.


```python
#Improving the above graph by showing varaition of size by payload mass and mapping the  0: Failure and 1: Sucess for more proper 
#epresentation on garph

# Mapping Class values to outcome
df['Outcome'] = df['Class'].map({0: 'Failure', 1: 'Success'})

# Define custom colors for Success and Failure
custom_colors = {'Success': 'green', 'Failure': 'red'}

# Create the scatterplot with custom bubble sizes
plt.figure(figsize=(35, 6))
sns.scatterplot(data=df, x="FlightNumber", y="PayloadMass", hue="Outcome", palette=custom_colors, size="PayloadMass", sizes=(50, 500), alpha=0.8)

# Set labels for x and y axis
plt.xlabel("Flight Number", fontsize=20)
plt.ylabel("Payload Mass (kg)", fontsize=20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title("Payload Mass by Flight Number and Launch Outcome", fontsize=22)

# Remove the top and right spines
sns.despine()

# Turn off vertical grid lines
plt.grid(False, axis='x')

# Create custom legend without 'PayloadMass'
custom_legend = [Line2D([0], [0], marker='o', color='w', markersize=10, markerfacecolor='red'),
                Line2D([0], [0], marker='o', color='w', markersize=10, markerfacecolor='green')]
legend_labels = ['Failure', 'Success']

# Add custom legend
plt.legend(custom_legend, legend_labels, title='Launch Outcome', fontsize=12, title_fontsize=12, loc='upper left')

# Set x-axis limits and ticks from 1 to 90
plt.xlim(-1, 95)
plt.xticks(range(1, 91))

plt.show()
```

We see that different launch sites have different success rates.  <code>CCAFS LC-40</code>, has a success rate of 60 %, while  <code>KSC LC-39A</code> and <code>VAFB SLC 4E</code> has a success rate of 77%.


```python
# Group the data by 'LaunchSite' and calculate the mean success rate
success_rates = df.groupby('LaunchSite')['Class'].mean()
success_rates_percentage = (success_rates * 100).round(2)
# success rates for each launch site
print("Success rates for each launch site:")
print(success_rates_percentage)
```

```python
# Sort success rates in descending order
success_rates_percentage = success_rates_percentage.sort_values(ascending=False)

plt.figure(figsize=(8, 3))
sns.barplot(x=success_rates_percentage.values, y=success_rates_percentage.index, palette='viridis', orient='h', saturation=0.75)
plt.title('Success Rates for Each Launch Site', fontsize=14)
plt.xlabel('Success Rate (%)', fontsize=12)
plt.ylabel('Launch Site', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlim(0, 100)

# Turn off vertical grid lines
plt.grid(False, axis='x')

for i, rate in enumerate(success_rates_percentage):
    plt.text(rate + 2, i, f'{rate:.2f}%', va='center', fontsize=12)

plt.tight_layout()
plt.show()
```

### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# HINT use groupby method on Orbit column and get the mean of Class column

# Calculate the mean of the Class column grouped by Orbit
orbit_success_rate = df.groupby('Orbit')['Class'].mean().reset_index()
orbit_success_rate = orbit_success_rate.sort_values(by='Class', ascending=False)
orbit_success_rate['Class'] *= 100

# Plot the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Orbit', y='Class', data=orbit_success_rate, palette='viridis')

# Add title and labels
plt.title('Success Rate by Orbit Type')
plt.xlabel('Orbit Type')
plt.ylabel('Success Rate')

# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend. 


The function will help you get the year from the date:


```python
# A function to Extract years from the date 
year=[]
def Extract_year(date):
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
    
```

```python
# # Extract year from the Date column
# df['Year'] = pd.to_datetime(df['Date']).dt.year

# # Calculate success rate for each year
# success_rate = df.groupby('Year')['Class'].mean()

```

```python
# Calculate success rate for each year (in percentage)
success_rate = df.groupby('Year')['Class'].mean() * 100

sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))
sns.lineplot(data=success_rate, color='royalblue', marker='o', markersize=8, markerfacecolor='darkblue')

plt.fill_between(success_rate.index, success_rate, color='lightblue', alpha=0.4)

plt.title('Success Rate Over the Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Success Rate (%)', fontsize=12)

plt.xticks(success_rate.index.astype(int), fontsize=12)

plt.ylim(-3, 100)

plt.grid(True, linestyle='--', alpha=0.7)

plt.annotate('', xy=(2014.8, 80), xytext=(2012, 5),
            arrowprops=dict(facecolor='red', shrink=0.05))
plt.annotate('Success Rate Increasing from  2013 ', xy=(2015, 60), xytext=(2011.7, 17), rotation=58, fontsize=14)

# Show the plot
plt.show()
```

You can observe that the success rate since 2013 kept increasing till 2017 (stable in 2014) and after 2015 it started increasing.
