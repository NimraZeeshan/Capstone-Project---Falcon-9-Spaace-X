## Task 1: Mark all launch sites on a map


First, let's try to add each site's location on a map using site's latitude and longitude coordinates


The following dataset with the name `spacex_launch_geo.csv` is an augmented dataset with latitude and longitude added for each site. 


```python
# Download and read the `spacex_launch_geo.csv`
spacex_csv_file = wget.download('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv')
spacex_df=pd.read_csv(spacex_csv_file)
```

Now, you can take a look at what are the coordinates for each site.


```python
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

```python
# Initial the map
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)
# For each launch site, add a Circle object based on its coordinate (Lat, Long) values. In addition, add Launch site name as a popup label


# Iterate over each row in the launch_sites DataFrame
for index, site in launch_sites_df.iterrows():
    # Extract latitude and longitude of the launch site
    lat, lon = site['Lat'], site['Long']
    
    # Create a circle for the launch site
    circle = folium.Circle(
        location=[lat, lon],
        radius=30,  # Adjust the radius as needed
        color='#001CED',  # Set the color of the circle
        fill=True  # Fill the circle
    )
    
    # Add a popup to the circle with the launch site name
    circle.add_child(folium.Popup(site['Launch Site']))
    
    # Create a marker for the launch site
    marker = folium.map.Marker(
        location=[lat, lon],
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12; color:#d35400;"><b>{site["Launch Site"]}</b></div>'
        )
    )
    
    # Add the circle and marker to the site map
    site_map.add_child(circle)
    site_map.add_child(marker)

# Display the updated site_map
site_map
```

```python
import folium
from folium import plugins
from folium.features import DivIcon
from geopy.distance import geodesic

# Coordinates of the Equator
equator_coords = (0, 0)

# Initialize the map
site_map1 = folium.Map(location=(0, 0), zoom_start=2)

# Launch sites data
launch_sites = {
    'CCAFS LC-40': (28.562302, -80.577356),
    'CCAFS SLC-40': (28.563197, -80.576820),
    'KSC LC-39A': (28.573255, -80.646895),
    'VAFB SLC-4E': (34.632834, -120.610745)
}

# Iterate over each launch site
for site, coords in launch_sites.items():
    # Calculate the distance from the launch site to the Equator
    distance_to_equator = geodesic(coords, equator_coords).kilometers
    # Calculate the distance from the Equator to the launch site
    distance_from_equator = geodesic(equator_coords, coords).kilometers
    
    # Create a PolyLine from the launch site to the Equator
    polyline_to_equator = folium.PolyLine(locations=[coords, equator_coords], color='blue', dash_array='5,5')
    site_map1.add_child(polyline_to_equator)
    
    # Add annotation showing the distance from the launch site to the Equator
    folium.Marker(location=coords, 
                  icon=DivIcon(icon_size=(150,36), 
                               icon_anchor=(0,0),
                               html=f'<div style="font-size: 10pt; color: blue;"><b>{distance_to_equator:.2f} km</b></div>')
                 ).add_to(site_map1)
    
    # Add a pointer for the launch site
    folium.Marker(location=coords, popup=site).add_to(site_map1)
    
    # Create a PolyLine from the Equator to the launch site
    polyline_from_equator = folium.PolyLine(locations=[equator_coords, coords], color='green', dash_array='5,5')
    site_map1.add_child(polyline_from_equator)
    
    
    
    # # Add annotation showing the distance from the Equator to the launch site
    # folium.Marker(location=equator_coords, 
    #               icon=DivIcon(icon_size=(150,36), 
    #                            icon_anchor=(0,0),
    #                            html=f'<div style="font-size: 10pt; color: green;"><b>{distance_from_equator:.2f} km</b></div>')
    #              ).add_to(site_map)

# Create a PolyLine representing the Equator
equator_line = folium.PolyLine(locations=[(-0.5, -180), (-0.5, 180)], color='red', weight=2)
site_map1.add_child(equator_line)

# Add annotation for the Equator line
folium.Marker(location=(-0.5, 0), 
              icon=DivIcon(icon_size=(150,36), 
                           icon_anchor=(0,0),
                           html='<div style="font-size: 10pt; color: red;"><b>Equator</b></div>')
             ).add_to(site_map1)

# Display the map
site_map1
```

# TASK 3: Calculate the distances between a launch site to its proximities


Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.


You can calculate the distance between two points on the map based on their `Lat` and `Long` values using the following method:


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163

# 3. Calculate the distance between the launch site and the coastline point using the provided calculate_distance function
coastline_lat = 28.5643
coastline_lon = -80.56817

launch_site_lat = launch_sites_df.iloc[0]['Lat']  # Example: Latitude of the first launch site
launch_site_lon = launch_sites_df.iloc[0]['Long']  # Example: Longitude of the first launch site

distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

print("Distance to the closest coastline:", distance_coastline, "km")

```

_TODO:_ After obtained its coordinate, create a `folium.Marker` to show the distance


```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# coordinate = [coastline_lat, coastline_lon]

# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#016C1D;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )

#Iam making function for the ease

def create_distance_marker(coordinate, distance):
    # Create a Marker object with the given coordinate and distance
    distance_marker = folium.Marker(
       coordinate,
       icon=DivIcon(
           icon_size=(20,20),
           icon_anchor=(0,0),
           html='<div style="font-size: 12; color:#016C1D;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
           )
       )
    return distance_marker

# Create a folium.Marker for the coastline label
coastline_label = folium.Marker(
    location=[coastline_lat, coastline_lon],
    icon=folium.Icon(color='blue'),
    popup='Coastline'
)

# Add the coastline label to the map
site_map.add_child(coastline_label)
```
